# 리스트 (immutable)

```python
a = [1,2,3,4,5,7]
print(a[2]+a[5])

# 리스트 함수들*********************************************************************************************
lst = [3, 1, 4, 2]
lst.append(5)        # [3, 1, 4, 2, 5]
lst.extend([6, 7])   # [3, 1, 4, 2, 5, 6, 7]
lst.insert(1, 10)    # [3, 10, 1, 4, 2, 5, 6, 7]
lst.remove(10)       # [3, 1, 4, 2, 5, 6, 7]
lst.sort()           # [1, 2, 3, 4, 5, 6, 7]
lst.reverse()        # [7, 6, 5, 4, 3, 2, 1]
lst.index(4)  # 3
lst.count(2)  # 1
len(lst)  # 4
sum(lst)  # 10
max(lst)  # 4
min(lst)  # 1
2 in lst  # True
5 in lst  # False


a[1]=4
print(a)

del a[1]
print(a)
a.append(9)
a.insert(0,10)

sorted(a, key=lambda x:x>5, reverse=True)       # 5초과인 숫자만 정렬하겠음 나머지는 디폴트값 정렬렬
# 핵심 요약
# key에 함수(또는 람다 함수)를 전달하면, 각 요소가 정렬되기 전에 해당 함수의 반환값을 기준으로 정렬됩니다.
# reverse를 추가하면 정렬 순서를 반대로 변경할 수 있습니다.
# sorted()와 함께 key 매개변수를 사용하면 다양한 복잡한 정렬 작업도 쉽게 처리할 수 있습니다! 😊
# 예시
names = ["Alice", "Bob", "Charlie", "Dave"]
result = sorted(names, key=len)
print(result)  # ['Bob', 'Dave', 'Alice', 'Charlie']

data = [(1, 3), (4, 1), (2, 2)]
result = sorted(data, key=lambda x: x[1])
print(result)  # [(4, 1), (2, 2), (1, 3)]

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

people = [Person("Alice", 30), Person("Bob", 25), Person("Charlie", 35)]
result = sorted(people, key=lambda p: p.age)
print([p.name for p in result])  # ['Bob', 'Alice', 'Charlie']


```

# 딕셔너리 (mutable)

```python
#
print(list(a.keys()))

# 딕셔너리 생성 방법
# 1. 중괄호 {} 사용

my_dict = {"name": "Alice", "age": 25}
# 2. dict() 함수 사용

my_dict = dict(name="Alice", age=25)

# 딕셔너리 메서드****************************************************************************************************
# keys(): 모든 키 반환
# values(): 모든 값 반환
# items(): (키, 값) 쌍을 반환
# update(): 딕셔너리 병합

my_dict = {"name": "Alice", "age": 25}
my_dict.keys()     # dict_keys(['name', 'age'])
my_dict.values()   # dict_values(['Alice', 25])
my_dict.items()    # dict_items([('name', 'Alice'), ('age', 25)])
my_dict.get('name') # 'Alice' ==> get을 쓰는 이유는 딕셔너리 외에 값이 들어가면 오류가 안뜨게 하기위해서 = None이 뜸뜸
my_dict.get('sdfkhllsdf', '없는 키값입니다')    # 결과값이 None일때 , 이후에 나오는것을 리턴함
'name' in my_dict   # True



# 기본 문법
# dict.update([other])
# other: 업데이트할 대상. 딕셔너리 또는 키-값 쌍(iterable)을 전달할 수 있습니다.
# 반환값: None (원본 딕셔너리를 직접 수정합니다).
# 사용 예제
# 1. 다른 딕셔너리 병합

dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}

dict1.update(dict2)
print(dict1)  # {'a': 1, 'b': 3, 'c': 4}
# dict2의 내용을 dict1에 병합합니다.
# 키 b의 값이 2에서 3으로 덮어씌워졌습니다.

# 리스트나 튜플로 전달
# update()는 키-값 쌍을 담은 iterable(리스트, 튜플 등)을 받을 수도 있습니다.

dict1 = {"a": 1}
dict1.update([("b", 2), ("c", 3)])
print(dict1)  # {'a': 1, 'b': 2, 'c': 3}
```

# 집합 (mutable)

```python
# set의 특징
# 중복을 허용하지 않음: 동일한 값이 여러 번 추가되더라도 하나의 값만 저장됩니다.
# 순서가 없음: 인덱스로 접근할 수 없습니다.
# 값들이 무작위 순서로 저장됩니다.
# 변경 가능(mutable): 요소를 추가하거나 삭제할 수 있습니다.
# 해시 가능한 값만 저장 가능: 리스트나 딕셔너리처럼 변경 가능한 객체는 저장할 수 없습니다.

# 기본 생성
my_set = {1, 2, 3}
print(my_set)  # {1, 2, 3}
my_set = set('Hello')   # {'o', 'e', 'H', 'l'}
# 이터러블(iterable)을 받아서 집합 생성.
my_set = set([1, 2, 2, 3])
print(my_set)  # {1, 2, 3}

# *****************************************************************************************************************
my_set = {1, 2, 3}
my_set.add(4)   # {1, 2, 3, 4} / add(x): 집합에 값 x를 추가. / 객체 한개만 가능
my_set.update([2, 3, 4])    # {1, 2, 3, 4} / update(iterable): 이터러블의 모든 요소를 집합에 추가.
my_set.remove(2)    # {1, 3, 4}] / remove(x): 값 x를 삭제. 값이 없으면 KeyError.
my_set.discard(4)  # 에러 없음 / discard(x): 값 x를 삭제. 값이 없으면 에러 발생하지 않음.
removed = my_set.pop() # 임의의의 값 출력력 pop(): 임의의 요소를 제거하고 반환.
print(my_set)   # pop하고 나머지 값 출력
my_set.clear()  # set() 모든 요소 제거.

set1 = {1, 2, 3}
set2 = {3, 4, 5}
set1 & set2   # {3}
set1.intersection(set2)  # {3}
set1 | set2  # {1, 2, 3, 4, 5} / set1 + set2
set1.union(set2)  # {1, 2, 3, 4, 5} / set1 + set2
set1.difference(set2)  # {1, 2} / set1 - set2
set1 - set2     # {1, 2} / set1 - set2
```
