```python
# setì˜ íŠ¹ì§•
# ì¤‘ë³µì„ í—ˆìš©í•˜ì§€ ì•ŠìŒ: ë™ì¼í•œ ê°’ì´ ì—¬ëŸ¬ ë²ˆ ì¶”ê°€ë˜ë”ë¼ë„ í•˜ë‚˜ì˜ ê°’ë§Œ ì €ì¥ë©ë‹ˆë‹¤.
# ìˆœì„œê°€ ì—†ìŒ: ì¸ë±ìŠ¤ë¡œ ì ‘ê·¼í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.
# ê°’ë“¤ì´ ë¬´ì‘ìœ„ ìˆœì„œë¡œ ì €ì¥ë©ë‹ˆë‹¤.
# ë³€ê²½ ê°€ëŠ¥(mutable): ìš”ì†Œë¥¼ ì¶”ê°€í•˜ê±°ë‚˜ ì‚­ì œí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
# í•´ì‹œ ê°€ëŠ¥í•œ ê°’ë§Œ ì €ì¥ ê°€ëŠ¥: ë¦¬ìŠ¤íŠ¸ë‚˜ ë”•ì…”ë„ˆë¦¬ì²˜ëŸ¼ ë³€ê²½ ê°€ëŠ¥í•œ ê°ì²´ëŠ” ì €ì¥í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.

# ê¸°ë³¸ ìƒì„±
my_set = {1, 2, 3}
print(my_set)  # {1, 2, 3}
my_set = set('Hello')   # {'o', 'e', 'H', 'l'}
# ì´í„°ëŸ¬ë¸”(iterable)ì„ ë°›ì•„ì„œ ì§‘í•© ìƒì„±.
my_set = set([1, 2, 2, 3])
print(my_set)  # {1, 2, 3}

# *****************************************************************************************************************
my_set = {1, 2, 3}
my_set.add(4)   # {1, 2, 3, 4} / add(x): ì§‘í•©ì— ê°’ xë¥¼ ì¶”ê°€. / ê°ì²´ í•œê°œë§Œ ê°€ëŠ¥
my_set.update([2, 3, 4])    # {1, 2, 3, 4} / update(iterable): ì´í„°ëŸ¬ë¸”ì˜ ëª¨ë“  ìš”ì†Œë¥¼ ì§‘í•©ì— ì¶”ê°€.
my_set.remove(2)    # {1, 3, 4}] / remove(x): ê°’ xë¥¼ ì‚­ì œ. ê°’ì´ ì—†ìœ¼ë©´ KeyError.
my_set.discard(4)  # ì—ëŸ¬ ì—†ìŒ / discard(x): ê°’ xë¥¼ ì‚­ì œ. ê°’ì´ ì—†ìœ¼ë©´ ì—ëŸ¬ ë°œìƒí•˜ì§€ ì•ŠìŒ.
removed = my_set.pop() # ì„ì˜ì˜ì˜ ê°’ ì¶œë ¥ë ¥ pop(): ì„ì˜ì˜ ìš”ì†Œë¥¼ ì œê±°í•˜ê³  ë°˜í™˜.
print(my_set)   # popí•˜ê³  ë‚˜ë¨¸ì§€ ê°’ ì¶œë ¥
my_set.clear()  # set() ëª¨ë“  ìš”ì†Œ ì œê±°.

set1 = {1, 2, 3}
set2 = {3, 4, 5}
set1 & set2   # {3}
set1.intersection(set2)  # {3}
set1 | set2  # {1, 2, 3, 4, 5} / set1 + set2
set1.union(set2)  # {1, 2, 3, 4, 5} / set1 + set2
set1.difference(set2)  # {1, 2} / set1 - set2
set1 - set2     # {1, 2} / set1 - set2
```
